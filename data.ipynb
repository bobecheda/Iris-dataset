```python
# Task 1: Load and Explore iris Dataset

# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# Set seaborn styling
sns.set(style="whitegrid")
```

```python
# Load the dataset with error handling
try:
    iris_raw = load_iris()
    iris_df = pd.DataFrame(iris_raw.data, columns=iris_raw.feature_names)
    iris_df['species'] = pd.Categorical.from_codes(iris_raw.target, iris_raw.target_names)
    print("‚úÖ Iris dataset loaded successfully.")
except Exception as e:
    print("‚ùå Failed to load dataset:", e)
```

```python
# Display the first few rows
print("\nüìå First 5 rows of the dataset:")
print(iris_df.head())
```

```python
# Check data types and missing values
print("\nüîç Data Types:")
print(iris_df.dtypes)

print("\nüîé Checking for missing values:")
print(iris_df.isnull().sum())
```

```python
# Clean the dataset (handle missing values if any)
try:
    iris_df = iris_df.dropna()  # Drop rows with missing values (if any)
    print("\n‚úÖ Missing values handled (dropped if present).")
except Exception as e:
    print("‚ùå Error during data cleaning:", e)
```

```python
# Task 2: Basic Data Analysis

# Basic statistics
print("\nüìä Summary statistics:")
print(iris_df.describe())

# Group by species and compute mean
print("\nüìö Mean values per species:")
species_means = iris_df.groupby('species').mean()
print(species_means)

# Observations
print("\nüí° Observation:")
print("Setosa tends to have smaller petal measurements compared to Virginica and Versicolor.")
```

```python
# Task 3: Data Visualization

# 1. Bar Chart: Average petal length per species
plt.figure(figsize=(8, 5))
species_means['petal length (cm)'].plot(kind='bar', color=['#FF9999', '#99FF99', '#9999FF'])
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 2. Histogram: Distribution of Sepal Length
plt.figure(figsize=(8, 5))
plt.hist(iris_df['sepal length (cm)'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# 3. Scatter Plot: Sepal Length vs Petal Length
plt.figure(figsize=(8, 5))
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
for species in iris_df['species'].unique():
    subset = iris_df[iris_df['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['petal length (cm)'],
                label=species, color=colors[species])
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 4. Line Plot: Simulated trend (Index vs Sepal Width)
plt.figure(figsize=(8, 5))
plt.plot(iris_df.index, iris_df['sepal width (cm)'], label='Sepal Width (cm)', color='purple')
plt.title('Sepal Width Across Samples')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Width (cm)')
plt.tight_layout()
plt.show()
```
